## Indexing

Existe tanto *loc* como *iloc*

```python
reviews.iloc[0]
```

Como ves, campeón, nos da todos los datos para la posición 0

Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.

This means that it's marginally easier to retrieve rows, and marginally harder to get retrieve columns. To get a column with iloc, we can do the following:

**Obtener todos los valores de una columna**

```python
reviews.iloc[:, 0]
```

**Valores del 0 al 3 (en posición, no en indice)**

```python
reviews.iloc[:3, 0]
```

Como ves, marca tres valores para la columna cero

```python
reviews.iloc[1:3, 0]
```

y aquí, los valores en la posición 2 y 3

```python
#Ahora una lista
reviews.iloc[[0, 1, 2], 0]
```

**Truco para valores finales**

```python
reviews.iloc[-5:, 0]
```

**Ejemplo con *loc***

```python
#Todas las filas, para esas columnas
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

## SELECCIÓN CONDICIONAL **(CLAVE)**

```python
reviews.country == 'Italy'
```

Más importante que llevarte un booleano, nos da **todas las filas que cumplen con una condición**

```python
reviews.loc[reviews.country == 'Italy']
```

**FUNCIÓN AND**

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

**FUNCIÓN OR**

```python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

**FUNCIÓN QUE INCLUYE VARIAS OPCIONES DE UNA LISTA**

```python
reviews.loc[reviews.country.isin(['Italy', 'France', 'US'])]
```

**FUNCIÓN QUE NOS DA VACÍOS Y NO VACÍOS**

```python
reviews.loc[reviews.price.notnull()]
```

```python
reviews.price.notnull()
```

Si te fijas al meter un booleano dentro de un loc te da los valores

## Agrupar datos *groupby()*

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

Aquí agrupamos por las puntuaciones y obtendremos la segunda definición de título, podríamos haber pillado cualquier otra cosa.

```python
reviews.groupby('points').apply(lambda df: df.title.iloc[1])
```

## Ajuste fino filipino

Ahora vamos a hacerlo más fino aún, agrupando por pais y provincia y sacaremos el mejor puto vino

```python
reviews.groupby(['country', 'province']).apply(lambda dr: dr.loc[dr.points.idxmax()])
```

otro groupby interesante será: 

```python
reviews.groupby(['country']).price.agg([len,min, max])
```

## Indices múltiples

```python
countries_reviewed=reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
countries_reviewed=countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

Para que descienda:

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_index()
```
